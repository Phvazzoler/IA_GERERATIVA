- Indroduçao

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

print("Versão do Tensorflow ",tf.__version__)
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images,train_labels),(test_images,test_labels) = fashion_mnist.load_data()
class_names= ['camiseta','calça','sueter','vestido','casaco','sandalha','camisa','tenis','bolsa','bota' ]
print(f"imagens de treino{train_images.shape}")
print(f"imagens de teste{train_images.shape}")

```

```python
plt.figure()
plt.imshow(train_images[678],cmap=plt.cm.binary)
plt.colorbar()
plt.grid(False)
plt.title(f"classe : { class_names[train_labels[678]]}")
plt.show()
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

CONSTRUINDO MLP

```python
modelo = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(10,activation="softmax")
])
modelo.summary()
```

```python
modelo.compile(optimizer="adam",
               loss="sparse_categorical_crossentropy",
               metrics=["accuracy"])
```

```python

print("INICIANDO O TREINAMENTO...")
history= modelo.fit(train_images,train_labels, epochs=10)
print("TREINAMENTO CONCLUIDOCOM SUCESSO!!")
```

```python
test_loss,test_acc = modelo.evaluate(test_images,test_labels,verbose=2)
print(f"\nAcuracia nos dados de teste:{test_acc:.4f}({(test_acc*100):.2f}%)!")
```

```python
predictions = modelo.predict(test_images)
imagem_idx = 100

classe_prevista= np.argmax(predictions[imagem_idx])
classe_real = test_labels[imagem_idx]
print(f"a rede acha que a imagem é : {class_names[classe_prevista]}")
print(f"O gabarito real é {class_names[classe_real]}")
plt.figure()
plt.imshow(test_images[imagem_idx], cmap=plt.cm.binary)
plt.title(f"Previsao da IA : {class_names[classe_prevista]} | Real : {class_names[classe_real]}")
```
